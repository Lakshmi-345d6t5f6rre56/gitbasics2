# Representing a sine wave
    a=sine(wt)
    a=amplitude
    w=frequency
    t=time

```python
import numpy as np
import matplotlib.pyplot as p
Time=np.arange(1,10,1)
Amplitude=np.sin(Time)
p.plot(Time,Amplitude)
p.grid()
p.title("sine wave a=sin(wt)")
p.xlabel("Time")
p.ylabel("Amplitude")
p.show()
```

```python
import pandas as pd
df=pd.read_csv("income.csv")
df
```

```python
import matplotlib.pyplot as p
years=df.columns[2::]
income=df.iloc[1,2:]
income
```

```python
import matplotlib.pyplot as p
years=df.columns[2::]
Alaska_income=df.iloc[1,2:]
p.plot(years,Alaska_income)
p.title("income of alaska from 2005-2013")
p.xlabel("years")
p.ylabel("income of alaska")
p.savefig("income of alaska from 2005-2013.jpg")
p.show()
```

## plot all state income from 2005-2013

```python
import matplotlib.pyplot as p
years=df.columns[-5::]
income=df.iloc[0:,2:]
p.plot(years,income)
p.title("income of all states from 2005-2013")
p.xlabel("years")
p.ylabel("states")
p.savefig("income of alaska from 2005-2013.jpg")
p.show()
```
